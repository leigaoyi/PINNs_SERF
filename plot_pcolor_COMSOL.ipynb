```python
# -*- coding: utf-8 -*-
"""
Created on Mon Mar 28 16:45:06 2022

@author: Administrator
绘制对比图
"""

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import cmath

file_idx = '0.4'
filename = './result/COMSOL_Rop_40/z{0}.csv'.format(file_idx)
df = pd.read_csv(filename, header=9, names=['x', 'y', 'z', 'Pz'])
## (x, y, z) u1, 20796*4

x = np.asarray(df['x'])
y = np.asarray(df['y'])
z = np.asarray(df['z'])
Pz = np.asarray(df['Pz'])

## 细网格？还是要转成极坐标的，在极坐标系下，把一维转成二维，再排序插值。

## 转xy坐标，为角度、半径坐标（极坐标），函数
## 

def xy_to_polar(x, y, Pz):
    r_xy = []
    theta_xy = []
    for i in range(len(x)):
        cn = complex(x[i], y[i])
        r_i, theta_i = cmath.polar(cn)
        r_xy.append(r_i)
        theta_xy.append(theta_i)
    r_xy = np.asarray(r_xy)    
    theta_xy = np.asarray(theta_xy)
    
    return (r_xy, theta_xy, Pz)
    
r, theta, Pz = xy_to_polar(x, y, Pz)
r_max = r.max()
```

```python
## 极坐标网格函数，先半径，后角度

grid_num = 200
r_sequence = np.linspace(0, r_max, grid_num)
theta_sequence = np.linspace(-np.pi, np.pi, grid_num)

Pz_grid = np.zeros((grid_num, grid_num))
Pz_grid_num = np.zeros((grid_num, grid_num))

r_slice = r_max/(grid_num-1)
theta_slice = 2*np.pi/(grid_num-1/2)

for point_idx in range(len(r)):
    r_idx = int((r[point_idx] - 0)/r_slice )
    theta_idx = int((theta[point_idx] + np.pi)/theta_slice )
    
    if theta_idx < 199:
        Pz_grid[r_idx, theta_idx+1] += Pz[point_idx]
        Pz_grid_num[r_idx, theta_idx+1] += 1
    
    Pz_grid[r_idx, theta_idx] += Pz[point_idx]
    Pz_grid_num[r_idx, theta_idx] += 1
    
    if theta_idx > 0:
        Pz_grid[r_idx, theta_idx-1] += Pz[point_idx]
        Pz_grid_num[r_idx, theta_idx-1] += 1
    
Pz_grid_mean = Pz_grid/(Pz_grid_num + 1e-10)    

Pz_grid_mean.min()
```
